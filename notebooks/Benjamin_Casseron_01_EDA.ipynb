```python
# Segmentation of columns
cols = [
    'FIRE_NAME', 'ICS_209_NAME', 'MTBS_FIRE_NAME', 'COMPLEX_NAME',
    'FIRE_YEAR', 'DISCOVERY_DATE', 'DISCOVERY_DOY', 'DISCOVERY_TIME',
    'CONT_DATE', 'CONT_DOY', 'CONT_TIME'
]

# Number of unique values per column
for col in cols:
    n_unique = df[col].nunique(dropna=True)
    print(f"{col}: {n_unique} unique values")
```

```python
#Conversion of DISCOVERY_DATE & CONT_DATE to DateTime format
df['DISCOVERY_DATE_DT'] = pd.to_datetime(df['DISCOVERY_DATE'], origin='julian', unit='D', errors='coerce')
df['CONT_DATE_DT'] = pd.to_datetime(df['CONT_DATE'], origin='julian', unit='D', errors='coerce')
```

```python
df[['DISCOVERY_DATE_DT', 'CONT_DATE_DT']].dtypes
```

```python
# Check if any containment occurs before discovery
df['DATE_INVERSION'] = df['CONT_DATE_DT'] < df['DISCOVERY_DATE_DT']
date_inversions = df[df['DATE_INVERSION'] == True]
print(f"Fires with containment before discovery: {len(date_inversions)}")
```

```python
# Option 1: Convert count_DOY to INT64 as 100% were mismatching
df['CONT_DOY'] = df['CONT_DOY'].astype('Int64')


# Check if FIRE_YEAR matches year of DISCOVERY_DATE_DT
df['YEAR_MATCH'] = df['DISCOVERY_DATE_DT'].dt.year == df['FIRE_YEAR']
print(f"Rows where FIRE_YEAR doesn't match DISCOVERY_DATE year: {(~df['YEAR_MATCH']).sum()}")

# Check if DOY matches the computed day-of-year
df['DISC_DOY_MATCH'] = df['DISCOVERY_DATE_DT'].dt.dayofyear == df['DISCOVERY_DOY']
df['CONT_DOY_MATCH'] = df['CONT_DATE_DT'].dt.dayofyear == df['CONT_DOY']
print(f"DISCOVERY_DOY mismatches: {(~df['DISC_DOY_MATCH']).sum()}")
print(f"CONT_DOY mismatches: {(~df['CONT_DOY_MATCH']).sum()}")
```

```python
# View the problematic rows
df_year_issue = df[~df['YEAR_MATCH']][['FIRE_YEAR', 'DISCOVERY_DATE_DT', 'FIRE_NAME']]
print(df_year_issue)
```

```python
df['DISCOVERY_DATE'] = pd.to_datetime(df['DISCOVERY_DATE'])
df['YEAR'] = df['DISCOVERY_DATE'].dt.year
```

```python
df["CONT_DATE"] = pd.to_datetime(df["CONT_DATE"], origin='julian', unit='D')
#Creating a column for the number of days
df['DURATION_DAYS'] = (df['CONT_DATE'] - df['DISCOVERY_DATE']).dt.days

# Filtering out the null values
df = df[df['DURATION_DAYS'] >= 0]

#Boxplot to show outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['DURATION_DAYS'])
plt.title("DurÃ©e des feux forestiers (jours) â€“ Boxplot complet")
plt.xlabel("DurÃ©e (jours)")
plt.show()
```

```python
# Calculate percentage distribution
fsc_distribution = df['FIRE_SIZE_CLASS'].value_counts(normalize=True).sort_index()

# Convert to DataFrame for plotting
fsc_df = fsc_distribution.reset_index()
fsc_df.columns = ['FIRE_SIZE_CLASS', 'Proportion']

# Convert to percentage for better readability
fsc_df['Percent'] = fsc_df['Proportion'] * 100

plt.figure(figsize=(8, 5))

# Plot barplot with inverted inferno palette
sns.barplot(x='FIRE_SIZE_CLASS', y='Percent', hue='FIRE_SIZE_CLASS', data=fsc_df, palette='hot')

plt.title("Fire Size Class Distribution (%)")
plt.xlabel("Fire Size Class")
plt.ylabel("Percentage (%)")

# Show percentage value above each bar
for i, pct in enumerate(fsc_df['Percent']):
    plt.text(i, pct + 0.5, f"{pct:.1f}%", ha='center')

plt.ylim(0, fsc_df['Percent'].max() + 5)  # Add some space above bars

plt.show()
```
